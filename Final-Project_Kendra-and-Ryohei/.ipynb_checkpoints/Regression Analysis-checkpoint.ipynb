```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn
import statsmodels.api as sm

%matplotlib inline
```

```python
P = (pd.read_csv('Project Proposal Data - Project Proposal Data.csv'))
```

```python
P
```

```python
P.index = pd.Index(P['country'])
del P['country']
P['beds_avail'] = P['beds_total'] * P['percent_availbeds'] / 100
P['hcw_avail'] = P['known_hcw'] * 0.8
P['beds_shortage'] = P['beds_peak'] - P['beds_avail']
P['hcw_shortage'] = P['hcw_required'] - P['hcw_avail']
P
```

```python
x1 = P[['beds_avail', 'hcw_avail', 'known_cases']]
y1 = P[['deaths_actual']]
x1 = sm.add_constant(x1)
model1 = sm.OLS(y1, x1)
result1 = model1.fit()
result1.summary()
```

```python
x2 = P[['beds_shortage', 'hcw_shortage', 'known_cases']]
y2 = P[['deaths_actual']]
x2 = sm.add_constant(x2)
model2 = sm.OLS(y2, x2)
result2 = model2.fit()
result2.summary()
```

```python
x3 = P[['labstaff', 'beds_avail', 'known_cases', ]]
y3 = P[['deaths_actual']]
x3 = sm.add_constant(x3)
model3 = sm.OLS(y3, x3)
result3 = model3.fit()
result3.summary()
```
